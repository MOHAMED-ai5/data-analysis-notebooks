```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from google.colab import files
uploaded = files.upload()
df = pd.read_csv("used_car_sales.csv")


print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df['Sold Date'] = pd.to_datetime(df['Sold Date'], errors='coerce')
placeholder_date = pd.Timestamp("1970-01-01")
df['Sold Date Placeholder'] = df['Sold Date'] == placeholder_date

df['Sold Price-$'] = pd.to_numeric(df['Sold Price-$'], errors='coerce').fillna(0)
df['Purchased Price-$'] = pd.to_numeric(df['Purchased Price-$'], errors='coerce').fillna(0)
df['Price-$'] = pd.to_numeric(df['Price-$'], errors='coerce').fillna(0)
df['Mileage-KM'] = pd.to_numeric(df['Mileage-KM'], errors='coerce')
```

```python
df['Sold'] = ((df['Sold Price-$'] > 0) & (~df['Sold Date Placeholder'])) | (df['Margin-%'] != 0)


print("\nPrice Stats:\n", df['Price-$'].describe())
print("\nMileage Stats:\n", df['Mileage-KM'].describe())
print("\nManufactured Year Stats:\n", df['Manufactured Year'].describe())
print("\nSold counts:\n", df['Sold'].value_counts())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Price-$'], bins=50, color="skyblue", edgecolor="black")
plt.title("Distribution of Car Price ($)")
plt.xlabel("Price ($)")
plt.ylabel("Count")
plt.show()

```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Mileage-KM'], bins=50, color="orange", edgecolor="black")
plt.title("Distribution of Mileage (KM)")
plt.xlabel("Mileage (KM)")
plt.ylabel("Count")
plt.show()

```

```python


top_manu = df['Manufacturer Name'].value_counts().head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top_manu.index, y=top_manu.values, palette="viridis")
plt.title("Top 10 Manufacturers")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.show()

```

```python
plt.figure(figsize=(5,4))
df['Sold'].value_counts().plot(kind="bar", color=["red","green"])
plt.title("Sold vs Unsold Cars")
plt.xlabel("Sold")
plt.ylabel("Count")
plt.show()
```

```python


plt.figure(figsize=(8,6))
plt.scatter(df['Mileage-KM'], df['Price-$'], s=8, alpha=0.6)
plt.xscale("log")
plt.title("Price vs Mileage (log scale)")
plt.xlabel("Mileage (KM)")
plt.ylabel("Price ($)")
plt.show()
```

```python
df['YearBin'] = pd.cut(df['Manufactured Year'],
                       bins=[0,1999,2004,2009,2014,2019,2024,2030],
                       labels=['<=1999','2000-2004','2005-2009','2010-2014','2015-2019','2020-2024','2025+'])
plt.figure(figsize=(10,6))
sns.boxplot(x="YearBin", y="Price-$", data=df, showfliers=False)
plt.title("Price by Manufactured Year Bins")
plt.xlabel("Manufactured Year")
plt.ylabel("Price ($)")
plt.show()
```

```python
corr_cols = ['Price-$','Mileage-KM','Manufactured Year','Purchased Price-$','Sold Price-$','Margin-%']
plt.figure(figsize=(6,5))
sns.heatmap(df[corr_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
issues = {
    "Placeholder sold dates (1970-01-01)": int(df['Sold Date Placeholder'].sum()),
    "Sold Price == 0 but Margin != 0": int(((df['Sold Price-$'] == 0) & (df['Margin-%'] != 0)).sum()),
    "Negative margins": int((df['Margin-%'] < 0).sum()),
    "Price Outliers <1% or >99%": int((df['Price-$'] < df['Price-$'].quantile(0.01)).sum()
                                     + (df['Price-$'] > df['Price-$'].quantile(0.99)).sum())
}
print("\nData Issues Found:")
for k,v in issues.items():
    print("-", k, ":", v)
```
