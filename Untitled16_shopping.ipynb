```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()
df = pd.read_csv('shopping_trends.csv')

print("First 5 rows of the dataset:")
print(df.head())


print("\nMissing values in each column:")
print(df.isnull().sum())

print("\nDataset Info:")
df.info()


print("\nDescriptive Statistics:")
print(df.describe())
```

```python
avg_purchase_by_category = df.groupby('Category')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
print("\nAverage Purchase Amount by Category:")
print(avg_purchase_by_category)
```

```python
popular_items = df['Item Purchased'].value_counts().head(10)
print("\nTop 10 Most Popular Items:")
print(popular_items)
```

```python
purchase_by_gender = df['Gender'].value_counts()
print("\nPurchase Count by Gender:")
print(purchase_by_gender)
```

```python
avg_rating_by_season = df.groupby('Season')['Review Rating'].mean()
print("\nAverage Review Rating by Season:")
print(avg_rating_by_season)
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numerical_cols.corr()
print("\nCorrelation Matrix:")
print(corr_matrix)
```

```python
sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_purchase_by_category.index, y=avg_purchase_by_category.values)
plt.title('Average Purchase Amount by Category')
plt.xlabel('Category')
plt.ylabel('Average Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
purchase_by_gender.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Season', y='Purchase Amount (USD)', data=df)
plt.title('Purchase Amount Distribution by Season')
plt.xlabel('Season')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
popular_items.plot(kind='bar')
plt.title('Top 10 Most Popular Items Purchased')
plt.xlabel('Item')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Purchase Amount (USD)', hue='Gender', data=df)
plt.title('Age vs Purchase Amount by Gender')
plt.show()
```
