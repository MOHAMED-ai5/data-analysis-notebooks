```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()


df = pd.read_csv("london_houses.csv")



print("Data format:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nGeneral information:")
print(df.info())
print("\nStatistical description:")
print(df.describe())

print("\nNumber of missing values ​​in each column:")
print(df.isnull().sum())
```

```python

print(df.columns.tolist())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price (£)"], bins=50, kde=True)
plt.title("Wine bottle distribution in London")
plt.xlabel("price(£)")
plt.ylabel("num of home")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Bedrooms", y="Price (£)", data=df)
plt.title("The relationship between the number of rooms and the price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Square Meters", y="Price (£)", data=df)
plt.title("The relationship between space and price")
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_neigh = df.groupby("Neighborhood")["Price (£)"].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_neigh.index, y=top_neigh.values)
plt.xticks(rotation=45)
plt.title("Top 10 neighborhoods by average price")
plt.ylabel("Average price(£)")
plt.show()
```

```python
insights = []


mean_price = df["Price (£)"].mean()
insights.append(f"Average house price in London: {mean_price:,.0f} £ almost.")

```

```python



min_price = df["Price (£)"].min()
max_price = df["Price (£)"].max()
insights.append(f"Cheapest house price{min_price:,.0f} £The most expensive house{max_price:,.0f} £.")
```

```python
avg_bedrooms = df.groupby("Bedrooms")["Price (£)"].mean()
insights.append("Average price by number of rooms:\n" + str(avg_bedrooms))
```

```python
avg_neigh = df.groupby("Neighborhood")["Price (£)"].mean().sort_values(ascending=False).head(5)
insights.append("Top 5 most expensive neighborhoods:\n" + str(avg_neigh))

print("\n====== Insights ======")
for i, ins in enumerate(insights, 1):
    print(f"{i}. {ins}")
```
